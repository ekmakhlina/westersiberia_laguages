```python
import pandas as pd
import re
import string
import collections
from nltk import word_tokenize, wordpunct_tokenize
import itertools
```

```python
with open('corpus_text.txt') as f:
    text_lines = f.read().splitlines()
```

```python
def make_all_sentence_df(text_lines):
    all_sentences = []
    all_morphemes = []
    all_glosses = []
    all_translations = []
    j = 0
    for line in text_lines:
        if j == 0:
            all_sentences.append(line)
        if j == 1:
            all_morphemes.append(line)
        if j == 2:
            all_glosses.append(line)
        if j == 3:
            all_translations.append(line)
        if j == 4:
            j = -1
        j += 1    
    all_sentences_df = pd.DataFrame(all_sentences)
    all_morphemes_df = pd.DataFrame(all_morphemes)
    all_glosses_df = pd.DataFrame(all_glosses)
    all_translations_df = pd.DataFrame(all_translations)
    frames = [all_sentences_df, all_morphemes_df, all_glosses_df, all_translations_df]
    result = pd.concat(frames, join='inner', axis=1)
    result.columns = ['sentence','morphemes','glosses', 'translation']
    return result
```

```python
result = make_all_sentence_df(text_lines)
result
```

```python
def string_into_tokens(s):
    ll = [[word_tokenize(w), ' '] for w in s.split()]
    s_list = list(itertools.chain(*list(itertools.chain(*ll))))
    return s_list
```

```python
def make_sentence_words_arr(all_sentences, i):
    wf_arr = []
    tokens = string_into_tokens(all_sentences['sentence'][i])
    morphemes = all_sentences['morphemes'][i].split()
    glosses = all_sentences['glosses'][i].split()
    j = 0
    off_start = 0
    sentence_index = 0
    for k in range(len(tokens)):
        if tokens[k] == ' ':
            off_start += len(tokens[k])
            continue
        elif tokens[k] in string.punctuation:
            wf_arr.append(collections.OrderedDict({
                'wf': tokens[k],
                'wtype': 'punct',
                'off_start': off_start,
                'off_end': off_start + len(tokens[k]),
                'next_word': sentence_index+1,
                'sentence_index': sentence_index
                #'ana': False
            }))
            sentence_index += 1
        else:
            wf_arr.append(collections.OrderedDict({
                'wf': tokens[k],
                'wtype': 'word',
                'off_start': off_start,
                'off_end': off_start + len(tokens[k]),
                'next_word': sentence_index+1,
                'sentence_index': sentence_index,
                'ana': [{'glosses': glosses[j], 'morphemes': morphemes[j]}]
                #"ana": [{"lex": "ojorsp", "gr.pos": "V", "gr.pers": "2", "gr.tense": "t1"}]
            }))
            j += 1
            sentence_index += 1
        off_start += len(tokens[k])
    return wf_arr
```

```python
all_sentences = []
```

```python
with open('corpus_text.txt') as f:
    text_lines = f.read().splitlines()
    all_sentences_df = make_all_sentence_df(text_lines)
    for i in range(len(all_sentences_df)):
        all_sentences.append(collections.OrderedDict({
            'text': all_sentences_df['sentence'][i],
            'words': make_sentence_words_arr(all_sentences_df, i),
            'lang': 0
        }))
```
